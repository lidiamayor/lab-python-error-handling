````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        incomplete = True
        while incomplete:
            try:
                inventory[product] = int(input(f"Enter the quantity of {product} available: ")) # ask user to input the quantity of each product
                if inventory[product] < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                incomplete = False
            except ValueError as error :
                print(f"Error: {error}")
    return inventory

def get_customer_orders(products, inventory):
    customer_orders = set() # empty set to store the customer orders
    act_products = products
    while True:
        try:
            quantity = int(input("Enter the number of customer orders: ")) # ask user to input the quantity of customer orders
            if quantity < 0:
                raise ValueError("Invalid quantity! Please enter a non-negative value.")
            break
        except ValueError as error:
            print(f"Error: {error}")


    while len(customer_orders) < quantity and quantity<=len(act_products): # check if the length of the customer orders is less than the quantity of customer_orders
        try:
            product = input(f"Please enter the product you want between {act_products}: ") # ask user to input the products they want
           
            if product not in act_products: # check if the input is in the list of products
                raise KeyError("This product is not available.")
            
            elif inventory[product]<=0:
                del inventory[product]
                act_products.remove(product)
                raise KeyError("The product is out of stock.")
            customer_orders.add(product) # add the product to the customer orders
            
        except KeyError as error:
            print(f"Error: {error}")
            

    print("Customer orders:")
    print(customer_orders)

    return customer_orders    

def update_inventory(customer_orders, inventory):
    inventory = {product: count - int(product in customer_orders) for product, count in inventory.items() if count - int(product in customer_orders) > 0}
    return inventory

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders) # count the total number of products ordered
    total_products = len(products) # count the total number of products available
    percentage = total_products_ordered/total_products*100 # calculate the percentage of products ordered
    order_status = (total_products_ordered, percentage) # tuple of total products ordered and percentage
    return order_status

def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Products Ordered:", order_statistics[1], "%")

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for key,value in inventory.items():
        print(f"{key}: {value}") # print the updated inventory

def calculate_total_price(customer_orders):
    prices = {}
    for product in customer_orders:
        while True:
            try:
                price = int(input(f"Enter the price of {product}: ")) # ask user to input the price of each product
                if price < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                prices[product] = price
                break
            except ValueError as error:
                print(f"Error. {error}")
    print("Total Price:", sum(prices.values()))
````

````python
def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"] # list of products
    inventory = initialize_inventory(products) # initialize the inventory
    print(inventory)
    print()
    customer_orders = get_customer_orders(products, inventory) # get the customer orders
    inventory = update_inventory(customer_orders, inventory) # update the inventory
    order_status = calculate_order_statistics(customer_orders, products) # calculate the order statistics
    print()
    print_order_statistics(order_status) # print the order statistics
    print()
    print_updated_inventory(inventory) # print the updated inventory
    print()
    calculate_total_price(customer_orders)
````
